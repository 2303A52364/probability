Generate a normal distribution of data points with a mean of 30 and a standard
deviation of 7. Plot a histogram for the generated data. Analyze the symmetry and bell-shaped curve of a normal distribution and visualize its Probability Density Function

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

# Generate random data from a normal distribution
mean = 30
std_dev = 7
num_samples = 1000
data = np.random.normal(mean, std_dev, num_samples)

# Plot histogram
plt.figure(figsize=(10, 5))
sns.histplot(data, bins=30, kde=True, color='blue', alpha=0.6)
plt.title("Histogram of Normally Distributed Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)

# Plot Probability Density Function (PDF)
x = np.linspace(mean - 3*std_dev, mean + 3*std_dev, 100)
pdf_values = norm.pdf(x, mean, std_dev)

plt.figure(figsize=(10, 5))
plt.plot(x, pdf_values, color='red', linewidth=2, label="PDF")
plt.title("Probability Density Function of Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.grid(True)

# Show plots
plt.show()
```

The normalized distribution for a person’s resting heart rate (RHR) has a mean of 50beats per minute (bpm) and a standard deviation of 10 bpm. Calculates and visualizesthe following probabilities for a randomly selected person having an RHR:

I. greater than or equal to 30 bpm.

II. II. less than or equal to 30 bpm.

III. III. between 30 and 50 bpm.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Given parameters
mean_rhr = 50  # Mean resting heart rate
std_dev_rhr = 10  # Standard deviation

# Compute probabilities using CDF
prob_greater_30 = 1 - norm.cdf(30, mean_rhr, std_dev_rhr)  # P(RHR ≥ 30)
prob_less_30 = norm.cdf(30, mean_rhr, std_dev_rhr)  # P(RHR ≤ 30)
prob_between_30_50 = norm.cdf(50, mean_rhr, std_dev_rhr) - norm.cdf(30, mean_rhr, std_dev_rhr)  # P(30 ≤ RHR ≤ 50)

# Print probabilities
print(f"Probability that RHR is ≥ 30 bpm: {prob_greater_30:.4f}")
print(f"Probability that RHR is ≤ 30 bpm: {prob_less_30:.4f}")
print(f"Probability that RHR is between 30 and 50 bpm: {prob_between_30_50:.4f}")

# Visualization
x = np.linspace(mean_rhr - 3*std_dev_rhr, mean_rhr + 3*std_dev_rhr, 1000)
pdf_values = norm.pdf(x, mean_rhr, std_dev_rhr)

plt.figure(figsize=(10, 5))
plt.plot(x, pdf_values, color='blue', label="PDF")

# Highlight regions
plt.fill_between(x, pdf_values, where=(x >= 30), color='green', alpha=0.3, label="≥ 30 bpm")
plt.fill_between(x, pdf_values, where=(x <= 30), color='red', alpha=0.3, label="≤ 30 bpm")
plt.fill_between(x, pdf_values, where=(30 <= x) & (x <= 50), color='orange', alpha=0.3, label="30-50 bpm")

plt.title("Probability Density Function of Resting Heart Rate")
plt.xlabel("Resting Heart Rate (bpm)")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()
```
