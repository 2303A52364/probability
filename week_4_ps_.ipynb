A factory produces products in batches, where each batch contains 0 to 8 defecƟve items dueto manufacturing variaƟons. The quality control team wants insights into defect rates.

 Generate a dataset represenƟng the number of defecƟve items per batch for N
batches.

 Compute and plot the PMF and CDF to analyze the probability of a batch having a specific number of defecƟve items

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set parameters
N = 1000  # Number of batches
max_defects = 8  # Defective items range from 0 to 8

# Generate synthetic defect data (assuming defects follow a random distribution)
np.random.seed(42)
data = np.random.randint(0, max_defects + 1, N)

# Compute PMF
pmf_values = {x: np.sum(data == x) / N for x in range(max_defects + 1)}

# Compute CDF
cdf_values = {x: np.sum(data <= x) / N for x in range(max_defects + 1)}

# Plot PMF
plt.figure(figsize=(10, 5))
plt.bar(pmf_values.keys(), pmf_values.values(), color='skyblue', alpha=0.7)
plt.xlabel("Number of Defective Items")
plt.ylabel("Probability")
plt.title("Probability Mass Function (PMF)")
plt.grid(True)

# Plot CDF
plt.figure(figsize=(10, 5))
plt.plot(list(cdf_values.keys()), list(cdf_values.values()), marker="o", linestyle="-", color="red")
plt.xlabel("Number of Defective Items")
plt.ylabel("Cumulative Probability")
plt.title("Cumulative Distribution Function (CDF)")
plt.grid(True)

# Show plots
plt.show()
```

A light bulb manufacturer produces large batches of bulbs, and each bulb has a 3% chance of being defecƟve. Suppose a quality inspector checks N bulbs (N ≥ 100):

i) What is the probability that exactly 5 bulbs are defecƟve?

ii) What is the probability that at least 8 bulbs are defecƟve?

iii) On average, how many defecƟve bulbs should the inspector expect?

```python
import scipy.stats as stats

# Parameters
N = 100  # Number of bulbs inspected
p = 0.03  # Probability of a bulb being defective

# Binomial distribution calculations
prob_5_defective = stats.binom.pmf(5, N, p)  # Probability of exactly 5 defective bulbs
prob_at_least_8_defective = 1 - stats.binom.cdf(7, N, p)  # Probability of at least 8 defective bulbs
expected_defective = N * p  # Expected number of defective bulbs

# Print results
print(f"Probability of exactly 5 defective bulbs: {prob_5_defective:.4f}")
print(f"Probability of at least 8 defective bulbs: {prob_at_least_8_defective:.4f}")
print(f"Expected number of defective bulbs: {expected_defective:.2f}")
```
