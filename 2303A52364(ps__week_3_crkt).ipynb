```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv("/content/cricket_dataset.csv")
data.head(50)


```

```python
def plot_histogram(data, column, bins=10):
    plt.figure(figsize=(10, 6))
    sns.histplot(data[column], bins=bins, kde=True, colour='blue')
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylable('Frequency')
    plt.show()
```

```python
def plot_scatter(data, x_column, y_cplumn):
  plt.figure(figsize=(10,6))
  sns.scatterplot(data=data, x=x_column, y=y_column, hue='Runs', palette='viridis')
  plt.title(f'Scatter Plot of {x_column} vs {y_column}')
  plt.xlabel(x_column)
  plt.ylabel(y_column)
  plt.show()
```

```python
def plot_bar_chart(data, x_column, y_column):
  plt.figure(figsize=(10,6))
  sns.barplot(data=data, x=x_column, y=y_column, palette='viridis')
  plt.title(f'Bar chart Plot of {x_column} vs {y_column}')
  plt.xlabel(x_column)
  plt.ylabel(y_column)
  plt.xticks(rotation=45)
  plt.show()
```

```python
file_path = ("/content/cricket_dataset.csv")
data = pd.read_csv(file_path)
```

```python
plot_histogram(data, 'Runs')
plot_histogram(data, 'Wickets')
```

```python
def plot_scatter(data, x_column, y_column): # Changed y_cplumn to y_column
  plt.figure(figsize=(10,6))
  sns.scatterplot(data=data, x=x_column, y=y_column, hue='Runs', palette='viridis')
  plt.title(f'Scatter Plot of {x_column} vs {y_column}')
  plt.xlabel(x_column)
  plt.ylabel(y_column)
  plt.show()
```
