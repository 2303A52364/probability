QuesƟon 1

A customer service center receives calls daily, with the number of calls ranging from 0 to 7 per
day. The management wants to study the distribuƟon of daily calls.
 Generate a dataset represenƟng the number of calls received per day for N days.
 Compute and visualize the PMF and CDF to analyze the probability of receiving a
specific number of calls on a given day.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate dataset
N = 1000  # Number of days
calls_per_day = np.random.randint(0, 8, N)  # Random calls between 0 and 7

# Compute PMF
unique, counts = np.unique(calls_per_day, return_counts=True)
pmf = counts / N  # Probability of each call count

# Compute CDF
cdf = np.cumsum(pmf)

# Plot PMF
plt.figure(figsize=(10, 5))
sns.barplot(x=unique, y=pmf, color='skyblue', edgecolor='black')
plt.xlabel("Number of Calls")
plt.ylabel("Probability")
plt.title("Probability Mass Function (PMF)")
plt.show()

# Plot CDF
plt.figure(figsize=(10, 5))
plt.plot(unique, cdf, marker='o', linestyle='-', color='red')
plt.xlabel("Number of Calls")
plt.ylabel("Cumulative Probability")
plt.title("Cumulative Distribution Function (CDF)")
plt.grid()
plt.show()
```

QuesƟon 2

A hospital receives emergency cases daily, and there is a 25% probability that a randomly
arriving paƟent requires immediate surgery. Suppose N paƟents visit the emergency
department in a day (N ≥ 100):
i) What is the probability that exactly 20 paƟents need surgery?
ii) What is the probability that at least 30 paƟents need surgery?
iii) On average, how many paƟents requiring surgery should the hospital expect?


```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Given parameters
N = 100  # Number of patients
p = 0.25  # Probability that a patient needs surgery

# i) Probability that exactly 20 patients need surgery
k_exact = 20
prob_exact_20 = stats.binom.pmf(k_exact, N, p)

# ii) Probability that at least 30 patients need surgery
prob_at_least_30 = 1 - stats.binom.cdf(29, N, p)  # 1 - P(X <= 29)

# iii) Average number of patients requiring surgery (expected value)
expected_surgery = N * p

# Output the results
print(f"Probability that exactly 20 patients need surgery: {prob_exact_20:.4f}")
print(f"Probability that at least 30 patients need surgery: {prob_at_least_30:.4f}")
print(f"On average, the hospital should expect {expected_surgery} patients requiring surgery.")

# Visualization - PMF and CDF

# PMF (Probability Mass Function)
x = np.arange(0, N + 1)
pmf_values = stats.binom.pmf(x, N, p)

# CDF (Cumulative Distribution Function)
cdf_values = stats.binom.cdf(x, N, p)

# Plotting PMF
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.bar(x, pmf_values, color='skyblue', label="PMF")
plt.title("Probability Mass Function (PMF)")
plt.xlabel("Number of Patients Needing Surgery")
plt.ylabel("Probability")
plt.xticks(np.arange(0, N + 1, 5))
plt.grid(True)

# Plotting CDF
plt.subplot(1, 2, 2)
plt.step(x, cdf_values, where='post', color='orange', label="CDF")
plt.title("Cumulative Distribution Function (CDF)")
plt.xlabel("Number of Patients Needing Surgery")
plt.ylabel("Cumulative Probability")
plt.xticks(np.arange(0, N + 1, 5))
plt.grid(True)

plt.tight_layout()
plt.show()
```
